# Pengertian Sistem persamaan linier

Sistem persamaan linear adalah sekumpulan persamaan linear yang memiliki satu atau lebih variabel dan saling berhubungan dalam satu sistem. Setiap persamaan dalam sistem ini berbentuk linear, yaitu variabel-variabelnya berpangkat satu dan tidak saling dikalikan.

Penyelesaian sistem persamaan linear dengan metode eliminasi Gauss didasarkan pada operasi baris dasar untuk mengubah sistem persamaan menjadi bentuk yang lebih sederhana, yaitu bentuk eselon baris. Berikut adalah langkah-langkahnya:

### Langkah-langkah Metode Eliminasi Gauss

#### 1. Menyusun Matriks Augmented

*    Dari sistem persamaan linear, buat matriks augmented (gabungan matriks koefisien dan matriks hasil).

#### 2. Mengubah Matriks ke Bentuk Eselon Baris

*    Gunakan operasi baris dasar (OBD) untuk mendapatkan nol di bawah elemen diagonal utama
* Tukar baris jika diperlukan.
* Kalikan satu baris dengan suatu bilangan bukan nol.
* Kurangkan atau tambahkan kelipatan suatu baris ke baris lain.
* Hasil akhirnya adalah bentuk segitiga atas (upper triangular form).

#### 3. Substitusi Mundur (Back Substitution)

* Setelah didapat bentuk eselon baris, lakukan substitusi mundur untuk menemukan nilai variabel.

#### contoh soal dan penyelesaiannya :
Diketahui sistem persamaan liniar

$\begin{aligned}
2x + y - z &= 8 \\
-3x - y + 2z &= -11 \\
-2x + y + 2z &= -3
\end{aligned}$


##### Langkah 1: Menyusun Matriks Augmented

$\begin{array}{ccc|c}
2 & 1 & -1 & 8 \\
-3 & -1 & 2 & -11 \\
-2 & 1 & 2 & -3
\end{array}$


##### Langkah 2: Eliminasi untuk Bentuk Eselon


* Buat elemen pertama kolom pertama menjadi 1 (opsional, tetapi tidak wajib).

* Nol-kan elemen di bawah elemen utama pertama (pivot):
  *  Baris 2 -> $R_2 + \frac{3}{2} R_1$

  *  Baris 3 -> $R_3 + R_1$

Sehingga diperoleh : 

$\begin{array}{ccc|c}
2 & 1 & -1 & 8 \\
0 & 0.5 & 0.5 & 1 \\
0 & 0 & -1 & 1
\end{array}$


##### Langkah 3: Subtitusi mundur

Dari persamaan terakhir:
$-1z = 1  z = -1$

Substitusi ke baris kedua:
$0.5y + 0.5(-1) = 1$

$0.5y - 0.5 = 1 \Rightarrow 0.5y = 1.5 \Rightarrow y = 3$



Substitusi ke baris pertama:
$\mathbf{2x + 3 - (-1) = 8}$
$\mathbf{2x + 3 + 1 = 8} \Rightarrow \mathbf{2x = 4} \Rightarrow \mathbf{x = 2}$

Sehingga solusi sistem persamaan adalah : 
$(x, y, z) = (2, 3, -1)$


### **Kesimpulan Metode Eliminasi Gauss**  

Metode eliminasi Gauss adalah teknik sistematis untuk menyelesaikan sistem persamaan linear dengan cara mengubah matriks augmented menjadi **bentuk eselon baris** menggunakan operasi baris dasar (OBD). Setelah bentuk eselon tercapai, solusi diperoleh melalui **substitusi mundur**.  

Metode ini memiliki keunggulan dalam kejelasan prosedur dan efisiensi komputasi, menjadikannya salah satu teknik fundamental dalam aljabar linear dan pemrograman numerik.

# Tugas 3 persamaan dengan 3 variabel dengan hasil 1,0,0

Diketahui sistem persamaan :
$\
\begin{aligned}
    6x - 3y - 2z &= 6 \\
    8x - 5y - 4z &= 8 \\
    2x - 7y - z &= 2
\end{aligned}$ 

Kita akan menghitung invers matriks **A** secara manual menggunakan metode **determinan dan adjoin**. Berikut langkah-langkahnya:  

---

### **Langkah 1: Tuliskan Matriks A**  
Diketahui matriks **A**:  

$\
A = \begin{bmatrix} 6 & -3 & -2 \\ 8 & -5 & -4 \\ 2 & -7 & -1 \end{bmatrix}$


Invers matriks $A^{-1}$ dapat dihitung dengan rumus:  

$A^{-1} = \frac{1}{\det(A)} \times \text{adj}(A)$


Di mana:

- $\det(A)$ adalah determinan matriks A
- $\text{adj}(A)$ adalah **adjoin**, yaitu transpose dari **matriks kofaktor**.







---

### **Langkah 2: Hitung Determinan \( A \)**  
Menggunakan ekspansi kofaktor pada baris pertama:

$\det(A) = 6 \left| \begin{matrix} -5 & -4 \\ -7 & -1 \end{matrix} \right| - (-3) \left| \begin{matrix} 8 & -4 \\ 2 & -1 \end{matrix} \right| - 2 \left| \begin{matrix} 8 & -5 \\ 2 & -7 \end{matrix} \right|$



Hitung determinan matriks 2Ã—2:

$\left| \begin{matrix} -5 & -4 \\ -7 & -1 \end{matrix} \right| = (-5)(-1) - (-4)(-7) = 5 - 28 = -23$

$\left| \begin{matrix} 8 & -4 \\ 2 & -1 \end{matrix} \right| = (8)(-1) - (-4)(2) = -8 + 8 = 0$

$\left| \begin{matrix} 8 & -5 \\ 2 & -7 \end{matrix} \right| = (8)(-7) - (-5)(2) = -56 + 10 = -46$



Sehingga:

$\det(A) = 6(-23) - (-3)(0) - 2(-46) = -138 + 0 + 92 = -46$


### **Langkah 3: Hitung Matriks Kofaktor**  
Matriks kofaktor dihitung dengan mencari **minoran** dari setiap elemen, lalu dikalikan dengan tanda **+/-** sesuai aturan.


Hitung elemen-elemen kofaktor:
$C_{11} = \left| \begin{matrix} -5 & -4 \\ -7 & -1 \end{matrix} \right| = -23$


$C_{12} = -\left| \begin{matrix} 8 & -4 \\ 2 & -1 \end{matrix} \right| = -0 = 0$

$C_{13} = \left| \begin{matrix} 8 & -5 \\ 2 & -7 \end{matrix} \right| = -46$

$C_{21} = -\left| \begin{matrix} -3 & -2 \\ 7 & -1 \end{matrix} \right| = -\left((-3)(-1) - (-2)(7)\right) = -(3 - 14) = 11$

$C_{22} = \left| \begin{matrix} 6 & -2 \\ 2 & -1 \end{matrix} \right| = (6)(-1) - (-2)(2) = -6 + 4 = -2$

$C_{23} = -\left| \begin{matrix} 6 & -3 \\ 2 & -7 \end{matrix} \right| = -\left((6)(-7) - (-3)(2)\right) = -(-42 + 6) = 36$

$C_{31} = \left| \begin{matrix} -3 & -2 \\ -5 & -4 \end{matrix} \right| = (-3)(-4) - (-2)(-5) = 12 - 10 = 2$

$C_{32} = -\left| \begin{matrix} 6 & -2 \\ 8 & -4 \end{matrix} \right| = -\left((6)(-4) - (-2)(8)\right) = -(-24 + 16) = 8$

$C_{33} = \left| \begin{matrix} 6 & -3 \\ 8 & -5 \end{matrix} \right| = (6)(-5) - (-3)(8) = -30 + 24 = -6$


Sehingga, matriks kofaktor:

$C = \begin{bmatrix} 
-23 & 0 & -46 \\ 
11 & -2 & 36 \\ 
2 & 8 & -6 
\end{bmatrix}$


### **Langkah 4: Hitung Matriks Adjoin**  
Adjoin dari matriks adalah **transpose** dari matriks kofaktor:
$C = \begin{bmatrix} 
-23 & 0 & -46 \\ 
11 & -2 & 36 \\ 
2 & 8 & -6 
\end{bmatrix}$


### **Langkah 5: Hitung Invers Matriks**  
Gunakan rumus:

$A^{-1} = \frac{1}{\det(A)} \times \operatorname{adj}(A)$


Karena $\ \det(A)$ = -46 \, maka:

$A^{-1} = \frac{1}{-46} \times 
\begin{bmatrix}
-23 & 11 & 2 \\
0 & -2 & 8 \\
-46 & 36 & -6
\end{bmatrix}$



Hasilnya:


$A^{-1} =
\begin{bmatrix}
0.5 & -0.2391 & -0.0435 \\
0 & 0.0435 & -0.1739 \\
1 & -0.7826 & 0.1304
\end{bmatrix}$





### **Langkah 6: Verifikasi dengan Perkalian $A^{-1} \cdot B$**  
Sekarang, kita kalikan:

$X = A^{-1} B$


Hasilnya adalah:

$X=
\begin{bmatrix}
1.0 \\
0.0 \\
-2.78 \times 10^{-16}
\end{bmatrix}$


Karena nilai $(-2.78 \times 10^{-16})$ sangat kecil (hampir nol), maka hasilnya benar-benar **(1,0,0)**.


## **Ringkasan Materi: Matrix Transformations (Halaman 225â€“238)**

### **1. Definisi Transformasi Matriks**
Transformasi matriks adalah suatu fungsi linear $T: \mathbb{R}^n \to \mathbb{R}^m$ yang dapat direpresentasikan sebagai perkalian matriks terhadap vektor:

$$
T(\vec{x}) = A\vec{x}
$$

di mana:
- $A$ adalah matriks ukuran $m \times n$,
- $\vec{x}$ adalah vektor kolom di $\mathbb{R}^n$,
- $T(\vec{x})$ adalah hasil transformasi vektor di $\mathbb{R}^m$.

---

### **2. Contoh Transformasi Matriks di $\mathbb{R}^2$**

Beberapa transformasi umum dalam $\mathbb{R}^2$ dapat direpresentasikan dengan matriks:

#### a. **Rotasi** sebesar $\theta$ terhadap titik asal:
$$
R_\theta =
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}
$$

#### b. **Refleksi:**
- Refleksi terhadap sumbu-$x$:
  $$
  \begin{bmatrix}
  1 & 0 \\
  0 & -1
  \end{bmatrix}
  $$
- Refleksi terhadap sumbu-$y$:
  $$
  \begin{bmatrix}
  -1 & 0 \\
  0 & 1
  \end{bmatrix}
  $$

#### c. **Skala (Stretching/Compression)**:
- Horizontal:
  $$
  \begin{bmatrix}
  k & 0 \\
  0 & 1
  \end{bmatrix}
  $$
- Vertikal:
  $$
  \begin{bmatrix}
  1 & 0 \\
  0 & k
  \end{bmatrix}
  $$

#### d. **Shear (Geser)**:
- Horizontal:
  $$
  \begin{bmatrix}
  1 & k \\
  0 & 1
  \end{bmatrix}
  $$
- Vertikal:
  $$
  \begin{bmatrix}
  1 & 0 \\
  k & 1
  \end{bmatrix}
  $$

---

### **3. Sifat-sifat Transformasi Linear**
Sebuah transformasi $T$ dikatakan **linear** jika memenuhi dua sifat berikut:

1. **Preservasi penjumlahan**:
   $$
   T(\vec{u} + \vec{v}) = T(\vec{u}) + T(\vec{v})
   $$

2. **Preservasi perkalian skalar**:
   $$
   T(c\vec{v}) = cT(\vec{v})
   $$

Jika kedua sifat tersebut terpenuhi, maka $T$ adalah **transformasi linear**, dan dapat direpresentasikan sebagai perkalian matriks.

---

### **4. Komposisi Transformasi**
Jika $T(\vec{x}) = A\vec{x}$ dan $S(\vec{x}) = B\vec{x}$, maka komposisi $T \circ S$ didefinisikan sebagai:

$$
T(S(\vec{x})) = A(B\vec{x}) = (AB)\vec{x}
$$

>**Catatan penting:** Perkalian matriks **tidak komutatif**, artinya $AB \ne BA$.

---

### **5. Identitas dan Matriks Nol**

- **Matriks identitas** $I$ tidak mengubah vektor:
  $$
  I\vec{x} = \vec{x}
  $$
- **Matriks nol** $0$ mengirim semua vektor ke nol:
  $$
  0\vec{x} = \vec{0}
  $$

---

### **6. Representasi Geometris**
Setiap matriks $2 \times 2$ menentukan bagaimana vektor basis standar berubah. Misalnya:

- Jika:
  $$
  A = \begin{bmatrix}
  a & b \\
  c & d
  \end{bmatrix}
  $$
- Maka:
  $$
  A\vec{e}_1 = \begin{bmatrix} a \\ c \end{bmatrix}, \quad A\vec{e}_2 = \begin{bmatrix} b \\ d \end{bmatrix}
  $$

Vektor hasil tersebut menunjukkan arah dan skala dari sumbu baru setelah transformasi.

---

### **7. Transformasi terhadap Unit Square**
Transformasi matriks terhadap **unit square** (persegi satuan) dapat divisualisasikan dengan melihat ke mana keempat titik sudut dikirim:

- Titik-titik unit square:
  $$
  (0, 0),\ (1, 0),\ (0, 1),\ (1, 1)
  $$
- Ditransformasi oleh $A$ menjadi:
  $$
  A\begin{bmatrix}0\\0\end{bmatrix},\ A\begin{bmatrix}1\\0\end{bmatrix},\ A\begin{bmatrix}0\\1\end{bmatrix},\ A\begin{bmatrix}1\\1\end{bmatrix}
  $$

---

### **8. Determinan dan Area**
- Untuk matriks $2 \times 2$:
  $$
  \det(A) = ad - bc
  $$
- Nilai absolut determinan menggambarkan **skala perubahan area** setelah transformasi.

---
