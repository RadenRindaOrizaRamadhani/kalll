# Pengertian Sistem persamaan linier

Sistem persamaan linear adalah sekumpulan persamaan linear yang memiliki satu atau lebih variabel dan saling berhubungan dalam satu sistem. Setiap persamaan dalam sistem ini berbentuk linear, yaitu variabel-variabelnya berpangkat satu dan tidak saling dikalikan.

Penyelesaian sistem persamaan linear dengan metode eliminasi Gauss didasarkan pada operasi baris dasar untuk mengubah sistem persamaan menjadi bentuk yang lebih sederhana, yaitu bentuk eselon baris. Berikut adalah langkah-langkahnya:

### Langkah-langkah Metode Eliminasi Gauss

#### 1. Menyusun Matriks Augmented

*    Dari sistem persamaan linear, buat matriks augmented (gabungan matriks koefisien dan matriks hasil).

#### 2. Mengubah Matriks ke Bentuk Eselon Baris

*    Gunakan operasi baris dasar (OBD) untuk mendapatkan nol di bawah elemen diagonal utama
* Tukar baris jika diperlukan.
* Kalikan satu baris dengan suatu bilangan bukan nol.
* Kurangkan atau tambahkan kelipatan suatu baris ke baris lain.
* Hasil akhirnya adalah bentuk segitiga atas (upper triangular form).

#### 3. Substitusi Mundur (Back Substitution)

* Setelah didapat bentuk eselon baris, lakukan substitusi mundur untuk menemukan nilai variabel.

#### contoh soal dan penyelesaiannya :
Diketahui sistem persamaan liniar

$\begin{aligned}
2x + y - z &= 8 \\
-3x - y + 2z &= -11 \\
-2x + y + 2z &= -3
\end{aligned}$


##### Langkah 1: Menyusun Matriks Augmented

$\begin{array}{ccc|c}
2 & 1 & -1 & 8 \\
-3 & -1 & 2 & -11 \\
-2 & 1 & 2 & -3
\end{array}$


##### Langkah 2: Eliminasi untuk Bentuk Eselon


* Buat elemen pertama kolom pertama menjadi 1 (opsional, tetapi tidak wajib).

* Nol-kan elemen di bawah elemen utama pertama (pivot):
  *  Baris 2 -> $R_2 + \frac{3}{2} R_1$

  *  Baris 3 -> $R_3 + R_1$

Sehingga diperoleh : 

$\begin{array}{ccc|c}
2 & 1 & -1 & 8 \\
0 & 0.5 & 0.5 & 1 \\
0 & 0 & -1 & 1
\end{array}$


##### Langkah 3: Subtitusi mundur

Dari persamaan terakhir:
$-1z = 1  z = -1$

Substitusi ke baris kedua:
$0.5y + 0.5(-1) = 1$

$0.5y - 0.5 = 1 \Rightarrow 0.5y = 1.5 \Rightarrow y = 3$



Substitusi ke baris pertama:
$\mathbf{2x + 3 - (-1) = 8}$
$\mathbf{2x + 3 + 1 = 8} \Rightarrow \mathbf{2x = 4} \Rightarrow \mathbf{x = 2}$

Sehingga solusi sistem persamaan adalah : 
$(x, y, z) = (2, 3, -1)$


### **Kesimpulan Metode Eliminasi Gauss**  

Metode eliminasi Gauss adalah teknik sistematis untuk menyelesaikan sistem persamaan linear dengan cara mengubah matriks augmented menjadi **bentuk eselon baris** menggunakan operasi baris dasar (OBD). Setelah bentuk eselon tercapai, solusi diperoleh melalui **substitusi mundur**.  

Metode ini memiliki keunggulan dalam kejelasan prosedur dan efisiensi komputasi, menjadikannya salah satu teknik fundamental dalam aljabar linear dan pemrograman numerik.

## **Ringkasan Materi: Matrix Transformations (Halaman 225â€“238)**

### **1. Definisi Transformasi Matriks**
Transformasi matriks adalah suatu fungsi linear $T: \mathbb{R}^n \to \mathbb{R}^m$ yang dapat direpresentasikan sebagai perkalian matriks terhadap vektor:

$T(\vec{x}) = A\vec{x}$

di mana:
- $A$ adalah matriks ukuran $m \times n$,
- $\vec{x}$ adalah vektor kolom di $\mathbb{R}^n$,
- $T(\vec{x})$ adalah hasil transformasi vektor di $\mathbb{R}^m$.

---

### **2. Contoh Transformasi Matriks di $\mathbb{R}^2$**

Beberapa transformasi umum dalam $\mathbb{R}^2$ dapat direpresentasikan dengan matriks:

#### a. **Rotasi** sebesar $\theta$ terhadap titik asal:
$R_\theta =
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}$

#### b. **Refleksi:**
- Refleksi terhadap sumbu-$x$:
  
  $\begin{bmatrix}
  1 & 0 \\
  0 & -1
  \end{bmatrix}$
  
- Refleksi terhadap sumbu-$y$:
  
  $\begin{bmatrix}
  -1 & 0 \\
  0 & 1
  \end{bmatrix}$

#### c. **Skala (Stretching/Compression)**:
- Horizontal:
  
  $\begin{bmatrix}
  k & 0 \\
  0 & 1
  \end{bmatrix}$
  
- Vertikal:
  
  $\begin{bmatrix}
  1 & 0 \\
  0 & k
  \end{bmatrix}$

#### d. **Shear (Geser)**:
- Horizontal:
  
  $\begin{bmatrix}
  1 & k \\
  0 & 1
  \end{bmatrix}$

- Vertikal:
  
  $\begin{bmatrix}
  1 & 0 \\
  k & 1
  \end{bmatrix}$

---

### **3. Sifat-sifat Transformasi Linear**
Sebuah transformasi $T$ dikatakan **linear** jika memenuhi dua sifat berikut:

1. **Preservasi penjumlahan**:
   $T(\vec{u} + \vec{v}) = T(\vec{u}) + T(\vec{v})$

2. **Preservasi perkalian skalar**:
   $T(c\vec{v}) = cT(\vec{v})$

Jika kedua sifat tersebut terpenuhi, maka $T$ adalah **transformasi linear**, dan dapat direpresentasikan sebagai perkalian matriks.

---

### **4. Komposisi Transformasi**
Jika $T(\vec{x}) = A\vec{x}$ dan $S(\vec{x}) = B\vec{x}$, maka komposisi $T \circ S$ didefinisikan sebagai:

$T(S(\vec{x})) = A(B\vec{x}) = (AB)\vec{x}$

>**Catatan penting:** Perkalian matriks **tidak komutatif**, artinya $AB \ne BA$.

---

### **5. Identitas dan Matriks Nol**

- **Matriks identitas** $I$ tidak mengubah vektor:
  $I\vec{x} = \vec{x}$
- **Matriks nol** $0$ mengirim semua vektor ke nol:
  $0\vec{x} = \vec{0}$

---

### **6. Representasi Geometris**
Setiap matriks $2\times2$ menentukan bagaimana vektor basis standar berubah. Misalnya:

- Jika:
  $A = \begin{bmatrix}
  a & b \\
  c & d
  \end{bmatrix}$
- Maka:
  $A\vec{e}_1 = \begin{bmatrix} a \\ c \end{bmatrix}, \quad A\vec{e}_2 = \begin{bmatrix} b \\ d \end{bmatrix}$

Vektor hasil tersebut menunjukkan arah dan skala dari sumbu baru setelah transformasi.

---

### **7. Transformasi terhadap Unit Square**
Transformasi matriks terhadap **unit square** (persegi satuan) dapat divisualisasikan dengan melihat ke mana keempat titik sudut dikirim:

- Titik-titik unit square:
  $(0, 0),\ (1, 0),\ (0, 1),\ (1, 1)$
- Ditransformasi oleh $A$ menjadi:
  $A\begin{bmatrix}0\\0\end{bmatrix},\ A\begin{bmatrix}1\\0\end{bmatrix},\ A\begin{bmatrix}0\\1\end{bmatrix},\ A\begin{bmatrix}1\\1\end{bmatrix}$

---

### **8. Determinan dan Area**
- Untuk matriks $2 \times 2$:
  $\det(A) = ad - bc$
- Nilai absolut determinan menggambarkan **skala perubahan area** setelah transformasi.

## Tugas 1

Misalkan $T: \mathbb{R}^2 \to \mathbb{R}^2$ didefinisikan oleh

$T(v_1, v_2) = (v_1 + v_2, v_1)$


Kita ingin membuktikan bahwa \( T \) adalah transformasi linier, yaitu memenuhi:

1. $T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})$ 
2. $T(c\mathbf{u}) = cT(\mathbf{u})$ 

dengan $\mathbf{u}, \mathbf{v} \in \mathbb{R}^2 \) dan \( c \in \mathbb{R}$.

---

**1. Pemeriksaan adisi**

Misalkan

$\mathbf{u} = (u_1, u_2), \quad \mathbf{v} = (v_1, v_2)$

maka

$\mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2)$

Hitung $T(\mathbf{u} + \mathbf{v})$:

$T(\mathbf{u} + \mathbf{v})$ = $T(u_1 + v_1, u_2 + v_2)$ = $((u_1 + v_1) + (u_2 + v_2), u_1 +v_1)$
= $(u_1 + u_2 + v_1 + v_2, u_1 + v_1)$

Hitung $T(\mathbf{u}) + T(\mathbf{v})$ :

$T(\mathbf{u}) = (u_1 + u_2, u_1)$

$T(\mathbf{v}) = (v_1 + v_2, v_1)$

$T(\mathbf{u}) + T(\mathbf{v}) = (u_1 + u_2 + v_1 + v_2, u_1 + v_1)$


Maka

$T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})$


Syarat adisi terpenuhi.



**2. Pemeriksaan homogenitas**

Ambil $c \in \mathbb{R}$ , maka

$c\mathbf{u} = (cu_1, cu_2)$


Hitung $T(c\mathbf{u})$:

$T(c\mathbf{u}) = T(cu_1, cu_2) = (cu_1 + cu_2, cu_1)$

$= c(u_1 + u_2, u_1)$

Di sisi lain:

$cT(\mathbf{u}) = c(u_1 + u_2, u_1) = (c(u_1 + u_2), cu_1)$


Maka

$T(c\mathbf{u}) = cT(\mathbf{u})$


Syarat homogenitas terpenuhi.



**Kesimpulan**

Karena kedua syarat terpenuhi, maka:

$\boxed{T \text{ adalah transformasi linier}}$

**Refleksi matriks** (reflection matrix) adalah **matriks transformasi** yang mencerminkan (memantulkan) titik-titik dalam ruang terhadap **sumbu, garis, bidang, atau titik tertentu**.

Refleksi mengubah arah satu atau lebih komponen vektor, seperti bayangan di cermin.

---

### âœ… **Repleksi pada 2D (dua dimensi)**:

$$
\begin{bmatrix}

\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

-y
\end{bmatrix}
$$

2. **Refleksi terhadap sumbu-y**:

$$
\begin{bmatrix}

\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}
$$

â†’ Membalik titik terhadap sumbu-y.

# Matrik Geser ( Shear Matrix)

**Shear matrix** atau **Matriks geser**. adalah jenis transformasi dalam **aljabar linear** dan **grafika komputer**, yang menggeser satu bagian dari objek ke arah tertentu tanpa mengubah volumenya.


Matriks shear "mendorong" titik-titik dalam suatu arah (biasanya horizontal atau vertikal), sedangkan titik-titik lain tetap relatif pada sumbu tertentu.

### Matrik Geser pada 2D:

Misalkan kita punya titik $(x, y)$. Jika kita terapkan **shear horizontal**, maka titiknya berubah menjadi:

$$
\begin{bmatrix}
1 & k \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
x + ky \\
y
\end{bmatrix}
$$

Artinya, sumbu-x digeser sebanding dengan nilai y (terjadi "kemiringan" arah horizontal).

Jika shear vertikal:

$$
\begin{bmatrix}
1 & 0 \\
k & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
x \\
y + kx
\end{bmatrix}
$$

### Aplikasi:

* Dalam **grafika komputer** untuk menciptakan efek distorsi
* Dalam **mekanika** untuk menggambarkan deformasi material
* Dalam **transformasi geometris**  computer vision

# Matrik Rotasi

**Matriks rotasi** adalah **matriks transformasi linier** yang digunakan untuk **memutar vektor atau objek** di ruang 2D, 3D, atau lebih tinggi **tanpa mengubah bentuk atau ukurannya**.

---


Matriks rotasi memutar titik atau objek **sekitar asal (0,0)** atau sumbu tertentu, dengan **sudut tertentu** (biasanya dalam radian atau derajat).

---

## âœ…  **Matriks Rotasi pada 2D**

Untuk memutar vektor sebesar sudut $\theta$ terhadap titik asal dalam bidang 2D:

$$
R(\theta) =
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}
$$

Jika kamu punya titik $(x, y)$, maka hasil rotasinya:

$$
\begin{bmatrix}
x' \\
y'
\end{bmatrix}
=
R(\theta) \cdot
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

---

## âœ…  **Matriks Rotasi pada 3D**

Di 3D, rotasi bisa dilakukan terhadap sumbu X, Y, atau Z.

1. **Rotasi terhadap sumbu X (dengan sudut $\theta$)**:

$$
R_x(\theta) =
\begin{bmatrix}
1 & 0 & 0 \\
0 & \cos\theta & -\sin\theta \\
0 & \sin\theta & \cos\theta
\end{bmatrix}
$$

2. **Rotasi terhadap sumbu Y**:

$$
R_y(\theta) =
\begin{bmatrix}
\cos\theta & 0 & \sin\theta \\
0 & 1 & 0 \\
-\sin\theta & 0 & \cos\theta
\end{bmatrix}
$$

3. **Rotasi terhadap sumbu Z**:

$$
R_z(\theta) =
\begin{bmatrix}
\cos\theta & -\sin\theta & 0 \\
\sin\theta & \cos\theta & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

---

# Komposisi transformasi

**Komposisi transformasi** (Composition of Transformations) adalah proses **menggabungkan dua atau lebih transformasi** geometris (seperti translasi, rotasi, refleksi, dan dilatasi) menjadi satu transformasi yang lebih kompleks.

Dalam komposisi transformasi, **transformasi pertama** diterapkan terlebih dahulu, dan kemudian **transformasi kedua** diterapkan pada hasil dari transformasi pertama. Jika lebih dari dua transformasi, proses ini dilakukan secara berurutan.


Misalnya, jika kita memiliki dua transformasi:

1. $T_1$ (seperti rotasi)
2. $T_2$ (seperti translasi)

Maka komposisi transformasi $T_1 \circ T_2$ berarti:

1. Terlebih dahulu **aplikasikan $T_2$** pada titik atau objek.
2. Setelah itu, **aplikasikan $T_1$** pada hasil transformasi sebelumnya.

### Matriks dan Komposisi:

Jika kita bekerja dengan **matriks transformasi**, komposisi dapat dilakukan dengan **perkalian matriks**. Misalnya, jika $A$ adalah matriks transformasi pertama dan $B$ adalah matriks transformasi kedua, maka komposisi transformasi tersebut adalah:

$$
C = A \times B
$$

Dengan catatan:

* Matriks $A$ diterapkan setelah $B$.
* Urutan perkalian matriks sangat penting, karena perkalian matriks **tidak komutatif** (yaitu $A \times B \neq B \times A$).

---

### Contoh Komposisi Transformasi:

Misalnya, kita ingin menggabungkan dua transformasi:

1. **Rotasi 90Â°** terhadap titik asal.
2. **Translasi (geser) 5 satuan ke kanan dan 3 satuan ke atas.**

Langkah-langkah:

1. **Rotasi**: Matriks rotasi 90Â°:

   $$
   R =
   \begin{bmatrix}
   0 & -1 \\
   1 & 0
   \end{bmatrix}
   $$

2. **Translasi**: Matriks translasi 5 satuan ke kanan dan 3 satuan ke atas:

   $$
   T =
   \begin{bmatrix}
   1 & 0 & 5 \\
   0 & 1 & 3 \\
   0 & 0 & 1
   \end{bmatrix}
   $$

Jika kita komposisikan translasi dan rotasi, kita harus menggabungkan matriks translasi dan rotasi ke dalam satu matriks transformasi. Hasil akhirnya akan memberikan rotasi **diikuti oleh translasi**.

---

* **Grafika komputer**: Untuk menggambar dan memanipulasi objek 3D (rotasi, translasi, zooming, dll.).
* **Robotika**: Untuk mengarahkan robot melalui serangkaian pergerakan.
* **Geometri**: Untuk mempelajari bentuk dan perubahan posisi dalam ruang.

# Matrik Translasi

**Matriks translasi** adalah matriks yang digunakan untuk **mentranslasi** (menggeser) titik atau objek dalam ruang. Translasi memindahkan titik dengan **jarak tertentu** di sepanjang sumbu tertentu, tanpa mengubah bentuk atau orientasi objek.

---

### âœ… **Matriks Translasi pada 2D**

Dalam ruang dua dimensi (2D), matriks translasi biasanya dituliskan sebagai:

$$
T =
\begin{bmatrix}
1 & 0 & tx \\
0 & 1 & ty \\
0 & 0 & 1
\end{bmatrix}
$$

* $tx$ adalah jarak translasi sepanjang sumbu **x** (horizontal).
* $ty$ adalah jarak translasi sepanjang sumbu **y** (vertikal).

Jika kita memiliki titik $(x, y)$, maka hasil translasi menjadi:

$$
\begin{bmatrix}
x' \\
y' \\
1
\end{bmatrix}
=
T \cdot
\begin{bmatrix}
x \\
y \\
1
\end{bmatrix}
$$

Dimana:

* $x' = x + tx$
* $y' = y + ty$

Jadi, titik $(x, y)$ akan digeser ke posisi baru $(x + tx, y + ty)$.

---

### âœ… **Matriks Translasi pada 3D**

Untuk translasi di ruang tiga dimensi (3D), matriks translasi dapat dituliskan sebagai:

$$
T =
\begin{bmatrix}
1 & 0 & 0 & tx \\
0 & 1 & 0 & ty \\
0 & 0 & 1 & tz \\
0 & 0 & 0 & 1
\end{bmatrix}
$$

* $tx$, $ty$, dan $tz$ adalah jarak translasi sepanjang sumbu **x**, **y**, dan **z**.

Jika kita memiliki titik $(x, y, z)$, maka hasil translasi menjadi:

$$
\begin{bmatrix}
x' \\
y' \\
z' \\
1
\end{bmatrix}
=
T \cdot
\begin{bmatrix}
x \\
y \\
z \\
1
\end{bmatrix}
$$

Dimana:

* $x' = x + tx$
* $y' = y + ty$
* $z' = z + tz$

---

### âœ… **Aplikasi Matriks Translasi:**

* **Grafika komputer**: Untuk menggeser objek atau gambar di layar.
* **Robotika**: Untuk memindahkan posisi lengan robot.
* **Geometri**: Untuk memindahkan titik atau bentuk dalam ruang tanpa merubah ukuran atau orientasi.

---




```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik objek (misalnya segitiga)
x = np.array([1, 2, 3])  # Koordinat X
y = np.array([1, 3, 2])  # Koordinat Y

# Matriks translasi (misalnya translasi 2 unit ke kanan dan 1 unit ke atas)
tx, ty = 2, 1
translation_matrix = np.array([[1, 0, tx],
                               [0, 1, ty],
                               [0, 0, 1]])

# Menambahkan baris 1 untuk koordinat homogen (x, y, 1)
coordinates = np.vstack((x, y, np.ones_like(x)))

# Menghitung hasil translasi
translated_coordinates = translation_matrix @ coordinates

# Titik-titik hasil translasi
x_translated = translated_coordinates[0, :]
y_translated = translated_coordinates[1, :]

# Plot objek asli dan hasil translasi
plt.figure(figsize=(6, 6))
plt.plot(x, y, 'bo-', label='Objek Asli')  # Titik objek asli (biru)
plt.plot(x_translated, y_translated, 'ro-', label='Objek Setelah Translasi')  # Titik objek setelah translasi (merah)

# Menambahkan label dan judul
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Matriks Translasi pada Objek 2D')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.legend()
plt.axis('equal')  # Agar skala X dan Y sama

plt.show()
```

## 1. **Definisi Eigenvalue dan Eigenvector**

Misalkan $A$ adalah matriks persegi berukuran $n \times n$. Jika terdapat **vektor bukan nol** $\vec{v}$ dan **skalar** $\lambda$ sehingga:

$$
A\vec{v} = \lambda \vec{v}
$$

maka:

* $\vec{v}$ adalah **eigenvector** dari matriks $A$
* $\lambda$ adalah **eigenvalue** (nilai eigen) yang terkait dengan $\vec{v}$

Artinya, operasi matriks $A$ terhadap $\vec{v}$ hanya mengubah **skala** vektor tersebut (memperbesar atau memperkecil), **tanpa mengubah arahnya**.

---

## 2. **Cara Mencari Eigenvalue dan Eigenvector**

### a. Mencari Eigenvalue

Untuk mencari eigenvalue dari matriks $A$, gunakan **persamaan karakteristik** berikut:

$$
\det(A - \lambda I) = 0
$$

Solusi dari persamaan ini adalah nilai-nilai $\lambda$ yang merupakan **eigenvalue** dari $A$.

### b. Mencari Eigenvector

Untuk setiap $\lambda$ yang ditemukan, cari vektor $\vec{v}$ dengan menyelesaikan:

$$
(A - \lambda I)\vec{v} = 0
$$

Solusi dari sistem homogen ini (selain $\vec{v} = 0$) adalah **eigenvector** yang sesuai dengan eigenvalue $\lambda$.

---

## 3. **Definisi Eigenvector Ortonormal**

### a. Apa Itu Eigenvector Ortonormal?

Sekelompok eigenvector ${\vec{v}\_1, \vec{v}\_2, \dots, \vec{v}\_n}$ disebut **ortonormal** jika:

1. **Ortogonal** (saling tegak lurus):

   $$
   \vec{v}_i \cdot \vec{v}_j = 0 \quad \text{untuk } i \ne j
   $$

2. **Norma satu** (panjang vektor = 1):

   $$
   \|\vec{v}_i\| = 1
   $$

Jika kedua syarat ini terpenuhi, maka vektor-vektor tersebut disebut **ortonormal**.

---

### b. Mengapa Eigenvector Ortonormal Penting?

* Dalam banyak aplikasi (seperti diagonalisis matriks simetris, PCA, dan transformasi orthogonal), eigenvector ortonormal sangat berguna.
* Eigenvector yang ortonormal dapat membentuk **basis ortonormal** dari ruang vektor.
* Ini menjadikan kumpulan eigenvector sebagai **satu kesatuan sistem yang stabil dan independen**, yang menyederhanakan banyak proses perhitungan dan analisis.

---


Kalimat **"eigenvector ortonormal menjadikan eigenvector satu kesatuan"** bisa dijelaskan secara matematis dan konseptual seperti ini dalam format **LaTeX untuk HackMD**:

---

## Eigenvector Ortonormal Menjadikan Eigenvector Satu Kesatuan

Jika kita memiliki satu set eigenvector yang **ortonormal**, maka vektor-vektor tersebut:

* **Saling bebas secara linear**
* **Tidak saling mempengaruhi (ortogonal)**
* **Mempunyai panjang (norma) yang sama, yaitu 1**

Dengan demikian, kumpulan vektor tersebut dapat membentuk **basis ortonormal** untuk ruang vektor.

### Secara matematis:

Misalkan terdapat $n$ eigenvector $\vec{v}\_1, \vec{v}\_2, \dots, \vec{v}\_n$ dari suatu matriks simetris $A$, maka dikatakan **ortonormal** jika:

$$
\vec{v}_i \cdot \vec{v}_j =
\begin{cases}
1, & \text{jika } i = j \\
0, & \text{jika } i \ne j
\end{cases}
$$

### Maknanya:

* Setiap vektor berdiri sendiri namun tetap **berkontribusi pada keseluruhan ruang**, tanpa redundansi.
* Ortonormalisasi menyatukan eigenvector menjadi **satu kesatuan sistem koordinat yang lengkap**, efisien, dan stabil secara numerik.

> **Kesimpulan:**
> Eigenvector ortonormal menjadikan kumpulan eigenvector **satu kesatuan sistem basis** yang utuh, terstruktur, dan siap digunakan dalam berbagai aplikasi matematika dan sains.
 Eigenvector ortonormal bukan hanya saling bebas (tidak saling pengaruh), tapi juga menyatu membentuk kerangka kerja (basis) yang kuat untuk representasi data, transformasi linear, dan pemahaman struktur matriks.
